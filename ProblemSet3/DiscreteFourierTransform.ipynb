```python
import numpy as np
import matplotlib.pyplot as plt
import math

dataLen = 100
x = np.linspace(-10, 10, dataLen)
y = np.exp(-0.5*x**2/(0.3**2))
data = np.zeros(shape=(2,dataLen))
data[0] = x
data[1] = y

def DFT(data):
    N = len(data[0])
    W = np.exp(2*np.pi*complex(0,1)/N)
    Wmat = np.zeros(shape=(N,N),dtype=complex)

    for n in range (N):
        for k in range(N):
            Wmat[n,k] = W**(n*k)
    H = np.dot(Wmat, data[1])
    Habs = np.absolute(H)
    return Habs

def FFT(data):
    N = len(data[0])
    W = np.exp(2*np.pi*complex(0,1)/N)
    return

plt.clf()

ydft = DFT(data)

ax1=plt.subplot(211)
plt.plot(data[0], data[1], 'k.')

ax2=plt.subplot(212)
plt.plot(data[0], ydft, 'r.')
plt.show()
```
